```python
import pandas as pd
import numpy as np
import scipy as sp
import plotly as py
from plotly.graph_objs import *
import plotly.figure_factory as ff
py.offline.init_notebook_mode(connected=True)
print('Python version ' + sys.version)
print('Pandas version ' + pd.__version__)
print('Matplotlib version ' + matplotlib.__version__)

df = pd.read_csv('/Users/cj/Downloads/test.csv')
print(df.columns)

table = ff.create_table(df)
py.offline.iplot(table, filename='jupyter/table1')

trace_BRONX = Bar(x=df.YEAR,
                  y=df.BRONX,
                  name='BRONX',
                  marker=dict(color='#ffcdd2'))

trace_BROOKLYN = Bar(x=df['YEAR'],
                y=df['BROOKLYN'],
                name='BROOKLYN',
                marker=dict(color='#A2D5F2'))

trace_MANHATTAN = Bar(x=df['YEAR'],
                y=df['MANHATTAN'],
                name='MANHATTAN',
                marker=dict(color='#59606D'))

trace_QUEENS = Bar(x=df['YEAR'],
                y=df['QUEENS'],
                name='QUEENS',
                marker=dict(color='#8cf795'))

trace_STATEN_ISLAND = Bar(x=df['YEAR'],
                y=df['STATEN ISLAND'],
                name='STATEN ISLAND',
                marker=dict(color='#c5ed15'))

data = [trace_BRONX, trace_BROOKLYN, trace_MANHATTAN, trace_QUEENS, trace_STATEN_ISLAND]
layout = Layout(title="Number of Crimes for each Borough in Each Year",
                xaxis=dict(title='YEAR'),
                yaxis=dict(title='Number of Crimes'))
fig = Figure(data=data, layout=layout)

py.offline.iplot(fig, filename='jupyter/styled_bar')
```
